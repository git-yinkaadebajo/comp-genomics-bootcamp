```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n_individuals = 100
n_snps = 500

genotype_matrix = np.random.randint(0, 3, size=(n_individuals, n_snps))

genotype_matrix.shape
```

```python
effect_sizes = np.random.normal(0, 0.1, n_snps)

prs = genotype_matrix @ effect_sizes

prs[:10]
```

```python
plt.hist(prs, bins=20)
plt.title("Simulated PRS Distribution")
plt.show()
```

```python
prs.shape
```

```python
# Manual PRS for first individual
manual_prs = np.sum(genotype_matrix[0] * effect_sizes)

manual_prs
```
